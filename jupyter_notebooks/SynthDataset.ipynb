```python
with_grid_sum = True
with_random_x_sum = True
with_grid_nonlinear = True
with_random_x_nonlinear = True
with_func_add_noise = False
with_z_add_noise = False
with_nesting = False
save_data_in_file = True
n_pts = 200
```

```python
def feature1(z1,z2,z3,z4,z5,z6,z7,z8,add_noise=False,add_noise_x=False):
    np.random.seed(13)   
    f1 = z1*z1+z2*z2+z3*z3+z4*z4+z5*z5+z6*z6+z7*z7+z8*z8
    return f1

def feature2(z1,z2,z3,z4,z5,z6,z7,z8,add_noise=False,add_noise_x=False):
    np.random.seed(3)   
    f2 = 100.0*np.log(z1)+np.log(np.abs(z2))+np.log(z3)+np.log(z4)+np.log(z5)+np.log(z6)+np.log(z7)+np.log(z8)
    return f2

def feature3(z1,z2,z3,z4,z5,z6,z7,z8,add_noise=False,add_noise_x=False):
    np.random.seed(10)
    f3 = z1*z1+z2*z2-2.0*z1*z2+z3*z3+z4*z4+z5*z5-z3*z4*z5+z6*z6+z7*z7+z8*z8-z6*z7*z8
    return f3

def feature4(z1,z2,z3,z4,z5,z6,z7,z8,add_noise=False,add_noise_x=False):
    np.random.seed(6) 
    f4 = 50.0*(np.sin(z1*z2)+np.sin(z3*z4)+np.sin(z5*z6)+np.sin(z7*z8))
    return f4
```

```python
np.random.seed(3)
if with_grid_sum:
    print('Generating functions with grid sampled independent variables')
    z1=np.linspace(0.5,10,n_pts) 
    z2=np.linspace(0.5,10,n_pts) 
    z3=np.linspace(0.5,10,n_pts) 
    z4=np.linspace(0.5,10,n_pts)  
    z5=np.linspace(0.5,10,n_pts) 
    z6=np.linspace(0.5,10,n_pts) 
    z7=np.linspace(0.5,10,n_pts)
    z8=np.linspace(0.5,10,n_pts)    

    f1=feature1(z1,z2,z3,z4,z5,z6,z7,z8)
    f2=feature2(z1,z2,z3,z4,z5,z6,z7,z8)
    f3=feature3(z1,z2,z3,z4,z5,z6,z7,z8)
    f4=feature4(z1,z2,z3,z4,z5,z6,z7,z8)
    f5=f1+f2+f3+f4 #np.sin(-(f1+f2)/(f3+f4))
    if save_data_in_file:
        functions=pd.DataFrame(data={'z1':z1,'z2':z2,'z3':z3,'z4':z4,'z5':z5,'z6':z6,'z7':z7,'z8':z8,
                                     'f1':f1,'f2':f2,'f3':f3,'f4':f4,'f5':f5})
        functions.to_csv('synthetic_data_gridSamples_'+str(n_pts)+'_sumf5.csv',index=False)
        
if with_random_x_sum:
    print('Generating functions with randomly sampled independent variables')
    z1=10.0*np.random.random(n_pts) 
    z2=10.0*np.random.random(n_pts) 
    z3=10.0*np.random.random(n_pts) 
    z4=10.0*np.random.random(n_pts) 
    z5=10.0*np.random.random(n_pts)
    z6=10.0*np.random.random(n_pts) 
    z7=10.0*np.random.random(n_pts) 
    z8=10.0*np.random.random(n_pts)    

    f1=feature1(z1,z2,z3,z4,z5,z6,z7,z8)
    f2=feature2(z1,z2,z3,z4,z5,z6,z7,z8)
    f3=feature3(z1,z2,z3,z4,z5,z6,z7,z8)
    f4=feature4(z1,z2,z3,z4,z5,z6,z7,z8)
    f5= f1+f2+f3+f4 #np.sin(-(f1+f2)/(f3+f4))
    if save_data_in_file:
        functions=pd.DataFrame(data={'z1':z1,'z2':z2,'z3':z3,'z4':z4,'z5':z5,'z6':z6,'z7':z7,'z8':z8,
                                     'f1':f1,'f2':f2,'f3':f3,'f4':f4,'f5':f5})
        functions.to_csv('synthetic_data_randomSamples_'+str(n_pts)+'_sumf5.csv',index=False)


fig,ax = plt.subplots(1,1,figsize=(6,5))
plt.plot(z1,f1, 'ob', markersize=5,markerfacecolor='None',markeredgewidth=2, label='f1')
plt.plot(z1,f2, 'or', markersize=5,markerfacecolor='None',markeredgewidth=2, label='f2')
plt.plot(z1,f3, 'ok', markersize=5,markerfacecolor='None',markeredgewidth=2, label='f3')
plt.plot(z1,f4, 'og', markersize=5,markerfacecolor='None',markeredgewidth=2, label='f4')
# plt.ylim([-30,30])
plt.xlabel('z1')
plt.ylabel('Nano scale functions')
plt.legend()

fig,ax = plt.subplots(1,1,figsize=(6,5))
plt.plot(z1,f5, 'ob', markersize=5,markerfacecolor='None',markeredgewidth=2, label='f5')
# plt.plot(z2,f5, 'or', markersize=5,markerfacecolor='None',markeredgewidth=2, label='f6')
# plt.plot(z3,f5, 'ok', markersize=5,markerfacecolor='None',markeredgewidth=2, label='f7')
# plt.plot(z4,f5, 'og', markersize=5,markerfacecolor='None',markeredgewidth=2, label='f8')
# plt.plot(z5,f5, 'o', markersize=5,markerfacecolor='None',markeredgewidth=2, label='f9')
# plt.plot(z6,f5, 'o', markersize=5,markerfacecolor='None',markeredgewidth=2, label='f10')
# plt.plot(z7,f5, 'o', markersize=5,markerfacecolor='None',markeredgewidth=2, label='f11')
# plt.plot(z8,f5, 'o', markersize=5,markerfacecolor='None',markeredgewidth=2, label='f12')
plt.xlabel('z1')
plt.ylabel('Target functions')
plt.legend()

if with_grid_nonlinear:
    print('Generating functions with grid sampled independent variables')
    z1=np.linspace(0.5,10,n_pts) 
    z2=np.linspace(0.5,10,n_pts) 
    z3=np.linspace(0.5,10,n_pts) 
    z4=np.linspace(0.5,10,n_pts)  
    z5=np.linspace(0.5,10,n_pts) 
    z6=np.linspace(0.5,10,n_pts) 
    z7=np.linspace(0.5,10,n_pts)
    z8=np.linspace(0.5,10,n_pts)    

    f1=feature1(z1,z2,z3,z4,z5,z6,z7,z8)
    f2=feature2(z1,z2,z3,z4,z5,z6,z7,z8)
    f3=feature3(z1,z2,z3,z4,z5,z6,z7,z8)
    f4=feature4(z1,z2,z3,z4,z5,z6,z7,z8)
    f5=np.sin(-(f1+f2)/(f3+f4))
    if save_data_in_file:
        functions=pd.DataFrame(data={'z1':z1,'z2':z2,'z3':z3,'z4':z4,'z5':z5,'z6':z6,'z7':z7,'z8':z8,
                                     'f1':f1,'f2':f2,'f3':f3,'f4':f4,'f5':f5})
        functions.to_csv('synthetic_data_gridSamples_'+str(n_pts)+'_nonlinearf5.csv',index=False)
        
if with_random_x_nonlinear:
    print('Generating functions with randomly sampled independent variables')
    z1=10.0*np.random.random(n_pts) 
    z2=10.0*np.random.random(n_pts) 
    z3=10.0*np.random.random(n_pts) 
    z4=10.0*np.random.random(n_pts) 
    z5=10.0*np.random.random(n_pts)
    z6=10.0*np.random.random(n_pts) 
    z7=10.0*np.random.random(n_pts) 
    z8=10.0*np.random.random(n_pts)    

    f1=feature1(z1,z2,z3,z4,z5,z6,z7,z8)
    f2=feature2(z1,z2,z3,z4,z5,z6,z7,z8)
    f3=feature3(z1,z2,z3,z4,z5,z6,z7,z8)
    f4=feature4(z1,z2,z3,z4,z5,z6,z7,z8)
    f5= np.sin(-(f1+f2)/(f3+f4))
    if save_data_in_file:
        functions=pd.DataFrame(data={'z1':z1,'z2':z2,'z3':z3,'z4':z4,'z5':z5,'z6':z6,'z7':z7,'z8':z8,
                                     'f1':f1,'f2':f2,'f3':f3,'f4':f4,'f5':f5})
        functions.to_csv('synthetic_data_randomSamples_'+str(n_pts)+'_nonlinearf5.csv',index=False)


fig,ax = plt.subplots(1,1,figsize=(6,5))
plt.plot(z1,f1, 'ob', markersize=5,markerfacecolor='None',markeredgewidth=2, label='f1')
plt.plot(z1,f2, 'or', markersize=5,markerfacecolor='None',markeredgewidth=2, label='f2')
plt.plot(z1,f3, 'ok', markersize=5,markerfacecolor='None',markeredgewidth=2, label='f3')
plt.plot(z1,f4, 'og', markersize=5,markerfacecolor='None',markeredgewidth=2, label='f4')
# plt.ylim([-30,30])
plt.xlabel('z1')
plt.ylabel('Nano scale functions')
plt.legend()

fig,ax = plt.subplots(1,1,figsize=(6,5))
plt.plot(z1,f5, 'ob', markersize=5,markerfacecolor='None',markeredgewidth=2, label='f5')
# plt.plot(z2,f5, 'or', markersize=5,markerfacecolor='None',markeredgewidth=2, label='f6')
# plt.plot(z3,f5, 'ok', markersize=5,markerfacecolor='None',markeredgewidth=2, label='f7')
# plt.plot(z4,f5, 'og', markersize=5,markerfacecolor='None',markeredgewidth=2, label='f8')
# plt.plot(z5,f5, 'o', markersize=5,markerfacecolor='None',markeredgewidth=2, label='f9')
# plt.plot(z6,f5, 'o', markersize=5,markerfacecolor='None',markeredgewidth=2, label='f10')
# plt.plot(z7,f5, 'o', markersize=5,markerfacecolor='None',markeredgewidth=2, label='f11')
# plt.plot(z8,f5, 'o', markersize=5,markerfacecolor='None',markeredgewidth=2, label='f12')
plt.xlabel('z1')
plt.ylabel('Target functions')
plt.legend()

# fig,ax = plt.subplots(1,1,figsize=(6,5))
# plt.plot(z1,f14, 'ob', markersize=5,markerfacecolor='None',markeredgewidth=2)
# plt.xlabel('z1')
# plt.ylabel('Coarse scale label')

# f5=feature5(z1,z2,z3,z4,z5)
# f6=feature6(z1,z2,z3,z4,z5)
# f7=feature7(z1,z2,z3,z4,z5)
# f8=feature8(z1,z2,z3,z4,z5)
# f9=feature9(z1,z2,z3,z4,z5)
# f11=feature11(z1,z2,z3,z4,z5)
# f12=feature12(z1,z2,z3,z4,z5)

# plt.plot(z1,f5, '-b', linewidth=3)
# plt.plot(z1,f6, '-r', linewidth=3)
# plt.plot(z1,f7, '-k', linewidth=3)
# plt.plot(z1,f8, '-g', linewidth=3)
# plt.plot(z1,f9, '-', linewidth=3)
# plt.plot(z1,f11, '-', linewidth=3)
# plt.plot(z1,f12, '-', linewidth=3)

# fig,ax = plt.subplots(1,1,figsize=(6,5))

# plt.plot(z1,f10, 'om', markersize=5,markerfacecolor='None',markeredgewidth=2)
# plt.plot(z1,f13, 'ob', markersize=5,markerfacecolor='None',markeredgewidth=2)
# f10=feature10(z1,z2,z3,z4,z5)
# f13=feature13(z1,z2,z3,z4,z5)
# plt.plot(z1,f10, '-r', linewidth=3)
# plt.plot(z1,f13, '-c', linewidth=3)
```
